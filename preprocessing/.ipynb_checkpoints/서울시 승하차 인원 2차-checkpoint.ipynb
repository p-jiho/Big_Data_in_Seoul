```python
import pandas as pd

from pyecharts.charts import Bar, Timeline
from pyecharts.globals import ThemeType
from pyecharts import options as opts
```

```python
data = pd.read_csv("../데이터(1차)/서울시_지하철_승하차_인원/202309.csv", encoding='cp949')

# "작업일자" column은 데이터 분석에 불필요한 열이라 제외
data = data.loc[:,~data.columns.isin(["작업일자"])] 
```

```python
line_lst = ['1호선','2호선','3호선','4호선','5호선','6호선','7호선','8호선','9호선','9호선2~3단계','경의선','경춘선',
 '공항철도 1호선','분당선','수인선','신림선','우이신설선','중앙선']
line = data.iloc[[i in line_lst for i in data.호선명],:]
line = data.loc[:,["호선명", "지하철역"] + [i for i in data.columns if "승차" in i] + [i for i in data.columns if "하차" in i]]
line.columns = [i[0:3] + "_승차" if "승차" in i else i for i in line.columns]
line.columns = [i[0:3] + "_하차" if "하차" in i else i for i in line.columns]
```

### 1호선

```python
line_1 = line[line["호선명"] == "1호선"]
line_1 = line_1.drop(["호선명"], axis = 1)

line_1_in = line_1.loc[:,["지하철역"] + [i for i in line_1.columns if "승차" in i]]
line_1_in.columns = [i[0:2] if "승차" in i else i for i in line_1_in.columns]
line_1_in = line_1_in.melt(id_vars = "지하철역", var_name = "시간", value_name = "승차인원")

line_1_out = line_1.loc[:,["지하철역"] + [i for i in line_1.columns if "하차" in i]]
line_1_out.columns = [i[0:2] if "하차" in i else i for i in line_1_out.columns]
line_1_out = line_1_out.melt(id_vars = "지하철역", var_name = "시간", value_name = "하차인원")
```

```python
tl = Timeline(init_opts=opts.InitOpts(theme=ThemeType.WHITE)).add_schema(is_auto_play=True)
for t in line_1_in["지하철역"].unique():
    tl.add((Bar(init_opts = opts.InitOpts(theme = ThemeType.WHITE))
            .add_xaxis(line_1_in["시간"].unique().tolist())
            .add_yaxis('승차', line_1_in[line_1_in["지하철역"] == t]["승차인원"].tolist(), color = "blue", gap = "0%")
            .add_yaxis('하차', line_1_out[line_1_out["지하철역"] == t]["하차인원"].tolist(), color = "red", gap = "0%")
            .set_global_opts(title_opts=opts.TitleOpts(title="1호선", subtitle = "{}".format(t)),
                              yaxis_opts=opts.AxisOpts(max_=250000))
           .set_series_opts(label_opts=opts.LabelOpts(is_show=False))),"{}".format(t))
    
tl.render_notebook()
```
